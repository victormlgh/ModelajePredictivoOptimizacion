```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data_url = 'smartcampuspucmm.csv'
df = pd.read_csv(data_url, encoding='unicode_escape')
print(df.shape)
df.head()
```

```python
date_format='%Y-%m-%d %H:%M:%S'
df['Fecha'] = pd.to_datetime(df['Fecha'], format=date_format)
dff = df.groupby(['Latitud','Longitud','Nombre',pd.Grouper(key='Fecha', freq='D')]).agg({'Valor':'mean'})
dff = dff.reset_index()
dff.head()
```

```python
dff.Nombre.unique()
```

```python
new_df = dff[['Latitud','Longitud','Fecha']]
new_df.head()
```

```python
for value in dff.Nombre.unique():
    temp_df = dff.loc[dff['Nombre']==value]
    temp_df = temp_df.drop(['Nombre'], axis=1)
    temp_df = temp_df.rename(columns = {'Valor':value})
    new_df = new_df.merge(temp_df, how='outer', indicator=False)

new_df.head()
```

```python
new_df = new_df.dropna()
new_df.head()
```

```python
new_df.to_csv('datosserie.csv', index=False)
```
